```python
scaler = StandardScaler()
```

```python
train2 = totaltrain.copy()
```

```python
lat_scaled = scaler.fit_transform(totaltrain.lat.values.reshape(-1, 1))
```

```python
lon_scaled = scaler.fit_transform(totaltrain.lon.values.reshape(-1, 1))
```

```python
d = pd.DataFrame({'x': [l for l in lat_scaled], 'y': [l for l in lon_scaled]})
```

```python
kmeans.fit(d)
```

```python
clusters = kmeans.predict(d)
```

```python
train2['cluster'] = clusters
```
